```python
# ,    DataFrame df

#    
columns_to_drop = ['W_plug', 'Vex', 'Vcurv', 'R', 'time_curv', 'Power_sum', 'B_zahv', 'Prod', 'Prod_expl', 'Prod_eff', 'Weight', 'Price']

#  
df_st = new1_df.drop(columns=columns_to_drop, inplace=False)

#  df    
```

```python

weights = {'St_Vex': 0.2,'St_Vcurv': 0.1,'St_R': 0.1,'St_Power_sum': -0.1,'St_B_zahv': 0.1,'St_Prod': 0.3, 'St_Prod_expl': 0.3, 'St_Prod_eff': 0.3, 'St_Weight': -0.1, 'St_Price': -0.2,'Plug': 1, 'Brush': 1, 'Air': 1, 'Vacuum': 1.1, 'Anti_ice': 1, 'Magnit': 1.1}

#    
selected_cols = weights.keys()

#    (    )
df_st['weighted_sum'] = df_st[selected_cols].fillna(0).multiply(weights, axis=1).sum(axis=1)

#     
df_sorted = df_st.sort_values(by='weighted_sum', ascending=False)
top_models = df_sorted.head()
```
